## 2. 경로에 해당하는 문자열 1개를 입력 받아 그 안에 디렉토리 경로명과 파일명을 분리하여 리스트로 반환하는 함수 div_path(s)를 작성하시오.

```python
def div_path(s):
    stackString = '' # 마지막 리스트 요소를 제외한 나머지를 받을 변수
    
    s = s.strip("'")
    s = s.split('/')
    s = s[1:]

    myList = [] # 최종 리스트를 담을 리스트
    for i in range(len(s)-1):
        stackString = stackString + '/' + s[i]
        
    myList.append(stackString)
    myList.append( s[len(s)-1])

    return myList

print("문자열을 입력하세요 : ")
myAddress = input() # 입력 받을 변수
print( div_path(myAddress) )
```

### '/home/chulsoo/test.txt' 를 입력받고 '를 제거해주기 위해 s.strip(" ' ")을 사용해주었다. s.split('/')이 실행되었을 때 s[0]에 '/'가 남아있기 때문에 s[1:]을 다시 s에 담아주어서 필요없는 값을 없애주었다. for문에 stackString + '/' + s[i]를 stackString에 담아주어 파일명을 제외한 경로명이 저장되도록 해주었고 이를 myList에 append 해주었으며, 마지막으로 파일명을 가르키던 s[len] - 1을 append하여 이를 반환해주었다. 

### 3. 두 개의 리스트를 인자로 받아서 그 두 개의 리스트에 대한 '합집합'을 반환하는 함수 list_union(lista, listb)를 작성하시오.

```python
def list_union(lista, listb):
    for i in listb:
        if ( i not in lista):
            lista.append(i)
   
    return sorted(lista)

a = [ 1, 2, 3 ]
b = [ 1, 2, 6, 4 ]
print ( list_union(a,b))
```

### a,b에 무작위 숫자로 구성된 리스트를 만들고 list_union함수를 호출한다. 만약 listb의 원소중 lista에 없는 원소가 있다면 이를 lista에 추가해주고 정렬된 lista를 반환하여 출력해주었다.

### 5. 두 개의 양의 정수를 인자로 받아서 해당 범위안에 있는 소수(prime number)의 리스트를 출력하는 함수 print_primenumber(a, b)를 작성하시오.

```python
def print_primenumber(a,b):
    pNumList = []
    for i in range(a,b+1):        
        count = 0
    
        for j in range(1,i+1):
            if i%j == 0:
                count += 1
        if count <= 2:
            pNumList.append(i)
        
    print("{} 이상 {} 이하 정수 중 소수 리스트: {}".format(a,b,pNumList))
            
print_primenumber(2,10)
```

### print_primenumber()함수를 2개의 정수 인자를 담아서 호출하였다. 이중 for문을 사용하여 첫 번째 for문은 받은 정수 2개 사이의 범위를 두 번째 for문은 받은 정수들을 나눠줄 인자에 해당하는 값들을 담았으며 만약 나눠떨어지면 count 값을 1씩 추가해주었다. 그 후, 소수인 값들만 pNumList에 담은 후 format을 활용하여 출력해주었다.

## [Incremental Project 문제]
### 웹 URL로 지정된 웹페이지를 문자열로 가져와 
### 1) 모든 HTML 태그를 제외한 순수 텍스트 문자열만을 걸러내고, 
### 2) 그 순수 텍스트 문자열 안에 존재하는 단어가 총 몇개인지 출력하는 프로그램을 작성하시오.


```python
import requests

req = requests.get('http://cse.koreatech.ac.kr')

source = req.text


source = source.strip()
string = ''
#print(source)
startRight = True
clearedList = []
for i in range( len(source)):
    if (source[i] == '<' ):
        startRight = False
        continue
    if ( source[i] == '>' and startRight == False):
        startRight = True
        continue
    if (source[i] == '\n' and startRight == True):
        startRight = False
        string += ' '
        continue
    if ( startRight == True ):
        string += source[i]

#print(string) 
clearedList = string.split()
print(clearedList)
print("문자열 안에 존재하는 단어 개수 : {}" .format(len(clearedList)))
```
